```python
import pandas as pd
```

```python

df = pd.read_csv("test.csv", sep=",", quotechar='"', engine="python", on_bad_lines="skip")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 1️⃣ Supprimer les lignes avec price = 0 ou < 1500
df = df[df["price"] >= 1500]
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python

# 2️⃣ Supprimer les doublons basés sur la colonne 'link'
df = df.drop_duplicates(subset=["link"], keep="first")
```

```python
df.info()
```

```python
# 3️⃣ Sauvegarder le CSV nettoyé
df.to_csv("products_clean.csv", index=False)
```
